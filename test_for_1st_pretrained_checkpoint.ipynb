```python
from qwerty_qwen2 import QwertyQwen2Model
from transformers import AutoTokenizer

model = QwertyQwen2Model.from_pretrained("/data/uchiha_ssd2/fengqi/241121_qwerty/231124_012656/checkpoint-4651")
tokenizer = AutoTokenizer.from_pretrained("/data/uchiha_ssd2/fengqi/241121_qwerty/231124_012656/checkpoint-4651")
```

```python
import os
from safetensors import safe_open
import torch
from typing import Dict

folder_path = "/data/uchiha_ssd2/fengqi/241121_qwerty/231124_012656/checkpoint-4651"
merged_weights: Dict[str, torch.Tensor] = {}
safetensors_files = [
    f for f in os.listdir(folder_path) 
    if f.endswith('.safetensors')
]

for file_name in safetensors_files:
    file_path = os.path.join(folder_path, file_name)
    with safe_open(file_path, framework="pt", device="cpu") as f:
        keys = f.keys()
        
        for key in keys:
            if key in merged_weights:
                print(f"警告: 键 {key} 在多个文件中出现,将使用文件 {file_name} 中的值")
            tensor = f.get_tensor(key)
            merged_weights[key] = tensor
```

```python
for key in model.state_dict().keys():
    if key.startswith('vision_model'):
        print(key)
        model.state_dict()[key] = merged_weights['model.' + key]
    if key.startswith('mm_projector'):
        print(key)
        model.state_dict()[key] = merged_weights['model.' + key]
```
